Question 1 : Explain the fundamental differences between DDL, DML, and DQL
commands in SQL. Provide one example for each type of command.


### Answer:

**DDL (Data Definition Language):**  
DDL commands are used to define and modify the structure of database objects, such as tables, indexes, and schemas. They do not manipulate the data itself but rather the schema of the database.  
Example: `CREATE TABLE students (id INT PRIMARY KEY, name VARCHAR(50));` - This creates a new table named 'students' with columns for id and name.

**DML (Data Manipulation Language):**  
DML commands are used to manipulate the data stored in the database tables. They allow inserting, updating, and deleting data.  
Example: `INSERT INTO students (id, name) VALUES (1, 'John Doe');` - This inserts a new row into the 'students' table.

**DQL (Data Query Language):**  
DQL commands are used to retrieve data from the database. The primary command is SELECT, which queries the database to fetch specific data based on conditions.  
Example: `SELECT name FROM students WHERE id = 1;` - This retrieves the name of the student with id 1 from the 'students' table.

Question 2 : What is the purpose of SQL constraints? Name and describe three common types
of constraints, providing a simple scenario where each would be useful.


### Answer:

The purpose of SQL constraints is to enforce rules on the data in a database to ensure data integrity, accuracy, and consistency. They prevent invalid data from being entered into the database and help maintain the quality of the data.

Three common types of constraints are:

1. **PRIMARY KEY Constraint:**  
   This ensures that each row in a table has a unique identifier and that the column(s) cannot contain NULL values. It uniquely identifies each record in the table.  
   Scenario: In a 'students' table, the 'student_id' column can be set as PRIMARY KEY to ensure no two students have the same ID and that every student has an ID.

2. **FOREIGN KEY Constraint:**  
   This establishes a relationship between two tables by ensuring that the value in one table matches a value in another table's primary key. It maintains referential integrity.  
   Scenario: In an 'enrollments' table, a 'student_id' column can be a FOREIGN KEY referencing the 'student_id' in the 'students' table, ensuring that enrollments only reference existing students.

3. **NOT NULL Constraint:**  
   This prevents a column from having NULL values, ensuring that every row must have a value for that column.  
   Scenario: In a 'products' table, the 'product_name' column can be set to NOT NULL to ensure that every product has a name and no products are created without a name.

Question 3 : Explain the difference between LIMIT and OFFSET clauses in SQL. How
would you use them together to retrieve the third page of results, assuming each page
has 10 records?

### Answer:

The LIMIT clause in SQL is used to restrict the number of rows returned by a query. It specifies the maximum number of records to fetch from the result set.

The OFFSET clause is used to skip a specified number of rows before starting to return rows from the query result. It is often used for pagination, allowing you to start retrieving data from a particular point in the result set.

To retrieve the third page of results with 10 records per page, you would use both LIMIT and OFFSET together. For the third page, you need to skip the first 20 records (2 pages × 10 records each) and then limit the results to 10 records.

Example query:
```sql
SELECT * FROM table_name ORDER BY some_column LIMIT 10 OFFSET 20;
```

This query will skip the first 20 rows and return the next 10 rows, which corresponds to the third page of results.

Question 4 : What is a Common Table Expression (CTE) in SQL, and what are its main
benefits? Provide a simple SQL example demonstrating its usage.


### Answer:

A Common Table Expression (CTE) in SQL is a temporary named result set that exists within the scope of a single SQL statement, such as SELECT, INSERT, UPDATE, or DELETE. It is defined using the WITH keyword and can be referenced multiple times within the same query.

The main benefits of CTEs include:

- **Improved readability:** CTEs allow you to break down complex queries into simpler, more understandable parts.

- **Reusability:** A CTE can be referenced multiple times in the same query, avoiding code duplication.

- **Recursion:** CTEs support recursive queries, which are useful for hierarchical data like organizational charts or bill of materials.

- **Modularity:** They help in organizing complex logic by separating different parts of the query.

Simple SQL example:
```sql
WITH high_salary_employees AS (
    SELECT employee_id, name, salary
    FROM employees
    WHERE salary > 50000
)
SELECT name, salary
FROM high_salary_employees
WHERE department = 'Engineering';
```

In this example, the CTE `high_salary_employees` filters employees with salaries over 50,000, and then the main query selects from this CTE to further filter by department.

Question 5 : Describe the concept of SQL Normalization and its primary goals. Briefly
explain the first three normal forms (1NF, 2NF, 3NF).


### Answer:

SQL Normalization is the process of organizing data in a database to reduce redundancy and improve data integrity. It involves decomposing tables into smaller, related tables and defining relationships between them to minimize data duplication and ensure that each piece of data is stored in only one place.

The primary goals of normalization are:
- Eliminate data redundancy (avoid storing the same data in multiple places).
- Ensure data integrity (maintain consistency and accuracy of data).
- Improve database efficiency (reduce storage space and improve query performance).
- Simplify database maintenance (make it easier to update, insert, and delete data without anomalies).

The first three normal forms are:

1. **First Normal Form (1NF):**  
   A table is in 1NF if it contains only atomic (indivisible) values, and there are no repeating groups or arrays. Each column must contain values of a single type, and each intersection of row and column must contain only one value.

2. **Second Normal Form (2NF):**  
   A table is in 2NF if it is in 1NF and all non-key attributes are fully functionally dependent on the entire primary key. This means that no partial dependencies exist; every non-key attribute must depend on the whole primary key, not just part of it.

3. **Third Normal Form (3NF):**  
   A table is in 3NF if it is in 2NF and there are no transitive dependencies. This means that non-key attributes should not depend on other non-key attributes. All non-key attributes must be directly dependent only on the primary key.

Question 7 : Generate a report showing CustomerName, Email, and the
TotalNumberofOrders for each customer. Include customers who have not placed
any orders, in which case their TotalNumberofOrders should be 0. Order the results
by CustomerName.

### Answer:

```sql code
SELECT 
    c.CustomerName, 
    c.Email, 
    COUNT(o.OrderID) AS TotalNumberofOrders
FROM Customers c
LEFT JOIN Orders o ON c.CustomerID = o.CustomerID
GROUP BY c.CustomerID, c.CustomerName, c.Email
ORDER BY c.CustomerName;
```

**Output:**

| CustomerName     | Email              | TotalNumberofOrders |
|------------------|--------------------|---------------------|
| Alice Wonderland | alice@example.com  | 3                   |
| Bob the Builder  | bob@example.com    | 2                   |
| Charlie Chaplin  | charlie@example.com| 1                   |
| Diana Prince     | diana@example.com  | 0                   |

**Explanation:**

- **LEFT JOIN:** Ensures all customers from the Customers table are included in the results, even those without orders
- **COUNT(o.OrderID):** Counts the number of orders for each customer. Customers with no orders will have a count of 0
- **GROUP BY:** Groups the results by customer to aggregate order counts
- **ORDER BY c.CustomerName:** Sorts the results alphabetically by customer name

Diana Prince appears with 0 orders because she is in the Customers table but has no matching records in the Orders table. The LEFT JOIN preserves her record with a count of 0.

Question 8 : Retrieve Product Information with Category: Write a SQL query to display the ProductName, Price, StockQuantity, and CategoryName for all products. Order the results by CategoryName and then ProductName alphabetically.

### Answer:

```sql
SELECT 
    p.ProductName,
    p.Price,
    p.StockQuantity,
    c.CategoryName
FROM Products p
INNER JOIN Categories c ON p.CategoryID = c.CategoryID
ORDER BY c.CategoryName, p.ProductName;
```

- **INNER JOIN:** Joins the Products table with the Categories table based on the CategoryID to fetch both product details and category names

Question 9 : Write a SQL query that uses a Common Table Expression (CTE) and a
Window Function (specifically ROW_NUMBER() or RANK()) to display the
CategoryName, ProductName, and Price for the top 2 most expensive products in
each CategoryName.

### Answer:

```sql
WITH RankedProducts AS (
    SELECT 
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC) AS PriceRank
    FROM Products p
    INNER JOIN Categories c ON p.CategoryID = c.CategoryID
)
SELECT 
    CategoryName,
    ProductName,
    Price
FROM RankedProducts
WHERE PriceRank <= 2
ORDER BY CategoryName, Price DESC;
```

**Output:**

| CategoryName | ProductName          | Price   |
|--------------|----------------------|---------|
| Apparel      | T-Shirt Casual       | 20.00   |
| Books        | SQL Handbook         | 45.50   |
| Books        | Novel : The Great SQL| 25.00   |
| Electronics  | Laptop Pro           | 1200.00 |
| Electronics  | Wireless Earbuds     | 150.00  |
| Home Goods   | Blender X            | 120.00  |
| Home Goods   | Coffee Maker         | 75.00   |

**Explanation:**

- **CTE (RankedProducts):** Defines a temporary named result set that:
  - Joins Products with Categories to get category information
  - Uses `ROW_NUMBER() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC)` to rank products within each category by price in descending order
  - PARTITION BY ensures each category gets its own ranking (1, 2, 3... for each category)
  - ORDER BY p.Price DESC ranks from most expensive to least expensive

- **Main Query:** Filters the CTE to show only products with PriceRank <= 2 (the top 2 most expensive per category)

- **Results:** For each category, displays the 2 most expensive products. Note that Apparel has only 1 product, so only 1 result appears for that category.

Question 10 : You are hired as a data analyst by Sakila Video Rentals, a global movie
rental company. The management team is looking to improve decision-making by
analyzing existing customer, rental, and inventory data.
Using the Sakila database, answer the following business questions to support key strategic
initiatives.
Tasks & Questions:
1. Identify the top 5 customers based on the total amount they’ve spent. Include customer
name, email, and total amount spent.
2. Which 3 movie categories have the highest rental counts? Display the category name
and number of times movies from that category were rented.
3. Calculate how many films are available at each store and how many of those have
never been rented.
4. Show the total revenue per month for the year 2023 to analyze business seasonality.
5. Identify customers who have rented more than 10 times in the last 6 months.

### Answer:

1. **Top 5 customers by total amount spent**

```sql
SELECT
    CONCAT(c.first_name, ' ', c.last_name) AS CustomerName,
    c.email,
    SUM(p.amount) AS TotalAmountSpent
FROM payment p
JOIN rental r ON p.rental_id = r.rental_id
JOIN customer c ON r.customer_id = c.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name, c.email
ORDER BY TotalAmountSpent DESC
LIMIT 5;
```

2. **Top 3 movie categories by rental count**

```sql
SELECT
    cat.name AS CategoryName,
    COUNT(*) AS RentalCount
FROM rental r
JOIN inventory i ON r.inventory_id = i.inventory_id
JOIN film f ON i.film_id = f.film_id
JOIN film_category fc ON f.film_id = fc.film_id
JOIN category cat ON fc.category_id = cat.category_id
GROUP BY cat.category_id, cat.name
ORDER BY RentalCount DESC
LIMIT 3;
```

3. **Films available at each store and how many have never been rented**

```sql
SELECT
    s.store_id,
    COUNT(DISTINCT i.film_id) AS FilmsAvailable,
    COUNT(DISTINCT CASE WHEN r.rental_id IS NULL THEN i.film_id END) AS FilmsNeverRented
FROM inventory i
JOIN store s ON i.store_id = s.store_id
LEFT JOIN rental r ON i.inventory_id = r.inventory_id
GROUP BY s.store_id
ORDER BY s.store_id;
```

4. **Total revenue per month for 2023**

```sql
SELECT
    DATE_FORMAT(p.payment_date, '%Y-%m') AS RevenueMonth,
    SUM(p.amount) AS TotalRevenue
FROM payment p
WHERE p.payment_date BETWEEN '2023-01-01' AND '2023-12-31'
GROUP BY RevenueMonth
ORDER BY RevenueMonth;
```

5. **Customers with more than 10 rentals in the last 6 months**

```sql
SELECT
    CONCAT(c.first_name, ' ', c.last_name) AS CustomerName,
    c.email,
    COUNT(r.rental_id) AS RentalCount
FROM rental r
JOIN customer c ON r.customer_id = c.customer_id
WHERE r.rental_date >= DATE_SUB(CURRENT_DATE, INTERVAL 6 MONTH)
GROUP BY c.customer_id, c.first_name, c.last_name, c.email
HAVING COUNT(r.rental_id) > 10
ORDER BY RentalCount DESC;
```

These queries use Sakila relationships to answer each business question clearly and efficiently.
